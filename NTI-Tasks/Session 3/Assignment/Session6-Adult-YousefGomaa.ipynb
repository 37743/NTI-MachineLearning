#### Preprocessing & Visualization

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.hist(figsize=(10, 10))
```

```python
import numpy as np

df[['age', 'fnlwgt']] = np.log1p(df[['age', 'fnlwgt']])
```

```python
df[['capital-gain', 'capital-loss']].hist(figsize=(10, 5))
```

```python
df.drop(columns=['capital-gain','capital-loss'], inplace=True)
```

```python
df.hist(figsize=(10, 10))
```

```python
value_counts = df['native-country'].value_counts()

categories_to_trim = value_counts[value_counts < 0.005 * len(df)].index

df = df[~df['native-country'].isin(categories_to_trim)]
```

```python
df['native-country'].value_counts().plot(kind='bar')
```

```python
df['native-country'].replace('?', df['native-country'].mode()[0], inplace=True)
```

```python
df['native-country'].value_counts().plot(kind='bar')
```

```python
df.dtypes
```

```python
for column in df.columns:
    if df[column].dtype != 'object':
        plt.subplots(figsize=(15, 5))
        plt.subplot(1, 2, 1)
        plt.title(f'Boxplot of {column}')
        plt.boxplot(df[column])
        q3 = df[column].quantile(0.75)
        q1 = df[column].quantile(0.25)
        df = df[df[column] <= q3+(q3-q1)*1.5]
        df = df[df[column] >= q1-(q3-q1)*1.5]
        plt.subplot(1, 2, 2)
        plt.title(f'Boxplot of {column}')
        plt.boxplot(df[column])
        plt.show()
```

```python
categorical_columns = df.select_dtypes(include=['object']).columns
print(categorical_columns)

some_categorical_columns = [categorical_columns[1],categorical_columns[2],categorical_columns[3]]

for i,column in enumerate(some_categorical_columns):
    value_counts = df[column].value_counts()
    plt.figure(figsize=(15, 5))
    plt.subplot(1, 3, i+1)
    plt.pie(value_counts, labels=value_counts.index, autopct='%1.2f%%', textprops={'fontsize': 8})
    plt.title(f'Pie Chart of {column}')
    plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

categorical_columns = df.select_dtypes(include=['object']).columns

for column in categorical_columns:
        df[column] = label_encoder.fit_transform(df[column])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df[['fnlwgt']] = scaler.fit_transform(df[['fnlwgt']])
```

##### SMOTE

```python
df['income'].value_counts()
```

```python
from imblearn.over_sampling import SMOTE

X = df.drop('income', axis=1)
y = df['income']

smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(X, y)

df = pd.concat([X_resampled, y_resampled], axis=1)
```
