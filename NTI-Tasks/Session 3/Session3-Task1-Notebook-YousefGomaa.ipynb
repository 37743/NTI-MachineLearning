```python
df = pd.read_csv('IBM Dataset.csv')
```

```python
df.info()
```

```python
missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values)
```

```python
duplicate_rows = df.duplicated()
print("Duplicate Rows:\n", duplicate_rows)
```

```python
data_types = df.dtypes
print("Data Types:\n", data_types)
```

```python
# Check for inconsistent categorical values
categorical_columns = df.select_dtypes(include=['object']).columns
for column in categorical_columns:
    unique_values = df[column].unique()
    print("Unique values in", column, ":\n", unique_values)
```

```python
df.describe().T
```

```python
df.hist(figsize=(20, 20))
```

```python
df.groupby(['JobRole'], as_index=False)['MonthlyIncome'].mean()
```

```python
df.groupby(['JobRole'], as_index=False)['MonthlyIncome'].mean().plot(kind='bar', x='JobRole', y='MonthlyIncome', title='Average Monthly Income by Job Role')
```

```python
from category_encoders import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler, StandardScaler

encoder = OneHotEncoder(cols=categorical_columns)
df_encoded = encoder.fit_transform(df)

minmax_scaler = MinMaxScaler()
df_minmax_scaled = minmax_scaler.fit_transform(df_encoded)

standard_scaler = StandardScaler()
df_standard_scaled = standard_scaler.fit_transform(df_encoded)
```
