```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_data = pd.read_csv('train_data.csv')
test_data = pd.read_csv('test_data.csv')
```

```python
train_data.head()
```

```python
train_data.describe()
```

```python
train_data['Age'] = train_data['Age']/365
```

```python
train_data['Age']
```

```python
for data in [train_data, test_data]:
    data.drop(['ID'], axis=1, inplace=True)
```

```python
for data in [train_data, test_data]:
    print(data.info())
```

```python
for data in [train_data, test_data]:
    print(data.isnull().sum())
```

```python
for data in [train_data, test_data]:
    data['Job title'].fillna('Unknown', inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

for data in [train_data, test_data]:
    for col in data.columns:
        if data[col].dtype == 'object':
            le = LabelEncoder()
            data[col] = le.fit_transform(data[col])
```

```python
train_data.head()
```

```python
for data in [train_data, test_data]:
    print(data.corr()['Is high risk'].sort_values(ascending=False))
```

```python
# remove outliers

for data in [train_data, test_data]:
    for col in data.columns:
        if data[col].dtype != 'object':
            q1 = data[col].quantile(0.25)
            q3 = data[col].quantile(0.75)
            iqr = q3 - q1
            data = data[(data[col] >= q1 - 1.5*iqr) & (data[col] <= q3 + 1.5*iqr)]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

X_train = train_data.drop(['Is high risk'], axis=1)
y_train = train_data['Is high risk']
X_test = test_data.drop(['Is high risk'], axis=1)
y_test = test_data['Is high risk']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#smote
from imblearn.over_sampling import SMOTE

smote = SMOTE()
X_train, y_train = smote.fit_resample(X_train, y_train)
```

```python
X_train.shape
```

```python
# Logistic Regression
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
log_reg_pred = log_reg.predict(X_test)

# KNN
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
knn_pred = knn.predict(X_test)

# SVM
svm = SVC(kernel='rbf', gamma=0.01)
svm.fit(X_train, y_train)
svm_pred = svm.predict(X_test)

from sklearn.metrics import accuracy_score

print('Logistic Regression:')
print('Accuracy:', accuracy_score(y_test, log_reg_pred))
print('KNN:')
print('Accuracy:', accuracy_score(y_test, knn_pred))
print('SVM:')
print('Accuracy:', accuracy_score(y_test, svm_pred))
```
