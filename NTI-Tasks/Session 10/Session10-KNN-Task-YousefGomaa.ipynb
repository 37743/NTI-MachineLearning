```python
from matplotlib import pyplot as plt
import numpy as np
import random
from sklearn.inspection import DecisionBoundaryDisplay
```

```python
import numpy as np
features = np.array([[1,0],[0,2],[1,1],[1,2],[1,3],[2,2],[3,2],[2,3]])
labels = np.array([0,0,0,0,1,1,1,1])
```

```python
# Plotting the points
plt.scatter(features[:,0], features[:,1], c=labels, cmap='coolwarm')
```

# KNN SKlearn

```python
from sklearn.neighbors import KNeighborsClassifier

accuracies = []

for k in range(3, 8, 2):
    clf = KNeighborsClassifier(n_neighbors=k)
    clf.fit(features, labels)

    plt.figure()
    DecisionBoundaryDisplay.from_estimator(clf, features, plot_method="pcolormesh", cmap='summer')
    plt.scatter(features[:,0], features[:,1], c=labels, cmap='Greens')
    plt.title(f"KNN with k={k}")
    plt.show()

    accuracy = clf.score(features, labels)
    accuracies.append(accuracy)
```

```python
plt.plot(range(3, 8, 2), accuracies)
plt.xlabel('k')
plt.ylabel('Accuracy')
plt.title('Accuracy vs k')
plt.show()

# print accuracies for each k
for k, accuracy in zip(range(3, 8, 2), accuracies):
    print(f"Accuracy for k={k}: {accuracy}")

best_k = 3 + 2 * np.argmax(accuracies)
print(f"\nThe best k is {best_k}")
```
